```python
import pandas as pd
import requests
import io
import os
```

## import file
change the output directory, the filename and the url according to the file you are working with <br>pay attention to use the **raw** url from github <br>Note: due to the token the url of the file changes, so it may have to be updated

```python
#outDir = os.chdir ("")
filename = "Area"
url = "https://raw.githubusercontent.com/gerlichlab/gerlich_general/master/data/hw1/Area.txt?token=AMCPTOTLVLJXKX5QWPMUI7S7IYDQA"
download = requests.get(url).content
df = pd.read_csv(io.StringIO(download.decode('utf-8')), sep = "\t")
```

```python
df.head()
```

tidy data: transform column variables into values

```python
df = pd.melt(df, id_vars = ['Absolute time'], value_vars=list(df.columns)[1:], var_name="trace")
df['trace'] = df['trace'].str.replace('Trace', '').astype('float')
df.to_csv(f"{filename}_tidy.csv", index=False)
```
